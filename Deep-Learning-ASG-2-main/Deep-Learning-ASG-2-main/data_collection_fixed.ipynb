```python
from google_play_scraper import Sort, reviews,reviews_all
import pandas as pd
```

```python
# initialization
app_name = 'com.disney.disneyplus' # found in app's url

rvws5, token5 = reviews(
    app_name,           # found in app's url
    lang='en',        # defaults to 'en'
    country='us',     # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # start with most recent
    count=200,        # batch size, set as 200 since 200 reviews per page
    filter_score_with=5,
)


app_reviews = rvws5
batch_num=1

rvws4, token4 = reviews(
    app_name,           # found in app's url
    lang='en',        # defaults to 'en'
    country='us',     # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # start with most recent
    count=200,        # batch size, set as 200 since 200 reviews per page
    filter_score_with=4,
)

app_reviews.extend(rvws4)

rvws3, token3 = reviews(
    app_name,           # found in app's url
    lang='en',        # defaults to 'en'
    country='us',     # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # start with most recent
    count=200,        # batch size, set as 200 since 200 reviews per page
    filter_score_with=3,
)

app_reviews.extend(rvws3)

rvws2, token2 = reviews(
    app_name,           # found in app's url
    lang='en',        # defaults to 'en'
    country='us',     # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # start with most recent
    count=200,        # batch size, set as 200 since 200 reviews per page
    filter_score_with=2,
)

app_reviews.extend(rvws2)

rvws1, token1 = reviews(
    app_name,           # found in app's url
    lang='en',        # defaults to 'en'
    country='us',     # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # start with most recent
    count=200,        # batch size, set as 200 since 200 reviews per page
    filter_score_with=1,
)

app_reviews.extend(rvws1)
```

```python
# Loop through the number of batches

for batch in range(50): # You can adjust the number of batches to collect more or less reviews
    rvws5, token5 = reviews( # store continuation_token
        app_name,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        # using token obtained from previous batch
        continuation_token=token5
    )
    app_reviews.extend(rvws5)
    batch_num +=1

for batch in range(50): # You can adjust the number of batches to collect more or less reviews
    rvws4, token4 = reviews( # store continuation_token
        app_name,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        # using token obtained from previous batch
        continuation_token=token4
    )
    app_reviews.extend(rvws4)
    batch_num +=1
    
for batch in range(50): # You can adjust the number of batches to collect more or less reviews
    rvws3, token3 = reviews( # store continuation_token
        app_name,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        # using token obtained from previous batch
        continuation_token=token3
    )
    app_reviews.extend(rvws3)
    batch_num +=1
    
for batch in range(50): # You can adjust the number of batches to collect more or less reviews
    rvws2, token2 = reviews( # store continuation_token
        app_name,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        # using token obtained from previous batch
        continuation_token=token2
    )
    app_reviews.extend(rvws2)
    batch_num +=1
    
for batch in range(50): # You can adjust the number of batches to collect more or less reviews
    rvws1, token1 = reviews( # store continuation_token
        app_name,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        # using token obtained from previous batch
        continuation_token=token1
    )
    app_reviews.extend(rvws1)
    batch_num +=1
```

```python
df = pd.DataFrame.from_records(app_reviews)
df
```

```python
df.to_csv('disney+_reviews.csv',index=False)
```
